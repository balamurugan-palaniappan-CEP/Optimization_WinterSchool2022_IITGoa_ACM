$\Large{\text{Recall:}}$


When $n=1$, that is when ${x} \in {\mathbb{R}}$, the real line, we can use basic calculus techniques to solve  $\min_{{x} \in {\mathbb{R}}} f({x})$. 




From calculus, we do the following:



1.   First we find the derivative (or) differential $f'({x})$ of the objective function  and then equate it to zero. Hence $f'({x}) = 0$ is solved to find some $\tilde{{x}}$ such that $f'(\tilde{{x}}) = 0$. 
2.   Then we check if $\tilde{{x}}$ is indeed a minimizer by checking if the double derivative of the objective function at $\tilde{{x}}$ is positive or not. Hence we check if $f''(\tilde{{x}}) > 0$. If this holds, we take $\tilde{x}$ as the minimizer. 




$\large{\text{Question:}}$ 

Can we extend the idea we saw in one dimension case to ${\mathbb{R}}^n$ where $n>1$ as well? 

___

Suppose we wish to extend the above idea. Let us assume that $\mathbf{x}=\begin{bmatrix}
x_1 \\ x_2 \\ \vdots \\ x_n
\end{bmatrix} \in {\mathbb{R}}^n$. 

Then we need to perform the following: 

1.   First we find the gradient of the objective function $\nabla f(\mathbf{x}) = \begin{bmatrix}
\frac{\partial f(\mathbf{x})}{\partial x_1} \\ \frac{\partial f(\mathbf{x})}{\partial x_2} \\ \vdots \\ \frac{\partial f(\mathbf{x})}{\partial x_n}
\end{bmatrix}$ and then equate it to zero vector. Hence $\nabla f(\mathbf{x}) = \mathbf{0}$ is solved to find some $\tilde{\mathbf{x}}$ such that $\nabla f(\tilde{\mathbf{x}}) = \mathbf{0}$. 
2.   Then we check if $\tilde{\mathbf{x}}$ is indeed a minimizer by checking if the Hessian matrix denoted by:
$
H_f(\tilde{\mathbf{x}}) = \nabla^2f(\tilde{\mathbf{x}})=\begin{bmatrix}
\frac{\partial^2 f(\tilde{\mathbf{x}})}{\partial x_1 \partial x_1} & \frac{\partial^2 f(\tilde{\mathbf{x}})}{\partial x_1 \partial x_2} & \ldots & \frac{\partial^2 f(\tilde{\mathbf{x}})}{\partial x_1 \partial x_n} \\
\vdots & \vdots & \ldots & \vdots \\
\frac{\partial^2 f(\tilde{\mathbf{x}})}{\partial x_n \partial x_1} & \frac{\partial^2 f(\tilde{\mathbf{x}})}{\partial x_n \partial x_2} & \ldots & \frac{\partial^2 f(\tilde{\mathbf{x}})}{\partial x_n \partial x_n} 
\end{bmatrix}
$
is $\textbf{positive definite}$ or not (by $\textbf{positive definite}$, we mean that $H_f$ matrix should have positive eigen values). 


$\large{\text{Recall:}}$ We call $\mathbf{x} \neq \mathbf{0}$ as an eigen vector of matrix $A$  with corresponding eigen value $\lambda$ when  

$A\mathbf{x}=\lambda \mathbf{x}$ holds.


## Notation: 
When $A$ is positive definite, we denote the positive definiteness property of $A$ by the notation: $A \succ 0$.  

#Quadratic functions 

One particular case where the above idea works is the minimization of a particular class of quadratic functions. 

Consider the problem $\min\limits_{\mathbf{x} \in {\mathbb{R}}^n} f(\mathbf{x}) = \frac{1}{2} {\mathbf{x}}^\top A {\mathbf{x}} + {\mathbf{b}}^\top {\mathbf{x}}+ c$, where $A$ is symmetric. Here $f$ is called a ${\color{red}{\textbf{quadratic}}}$ function. 

The gradient of $f$ is given by:

$
{\nabla} f(\mathbf{x}) = A\mathbf{x} + \mathbf{b}.  
$

The Hessian of $f$ is given by:

$
H_f(\mathbf{x}) = \nabla^2 f(\mathbf{x}) = A.
$

Thus if $\nabla f(\tilde{\mathbf{x}}) = A\tilde{\mathbf{x}} + \mathbf{b}=\mathbf{0}$ and if $H_f(\tilde{\mathbf{x}})=A$ is positive definite then we can certify that $\tilde{\mathbf{x}}$ is indeed a local minimizer. 

Further we have a nice closed form expression for the minimizer as $\tilde{x} = -A^{-1}\mathbf{b}$. ${\color{OrangeRed}{\textbf{(how?)}}}$

Quadratic function where the matrix $A$ is positive definite is called ${\color{orange}{\textbf{(strictly)}}}$ ${\color{blue}{\textbf{convex quadratic}}}$ function.

___

#Convex quadratic function

A quadratic function $f(\mathbf{x}) = \frac{1}{2} {\mathbf{x}}^\top A {\mathbf{x}} + {\mathbf{b}}^\top {\mathbf{x}}$ where $A \succeq 0$ (or $A$ is positive semi-definite) is called convex quadratic function. 

$\textbf{Note:}$ $A$ is positive semi-definite if all its eigen values are non-negative.
___

#Example

As an example of a (strictly) convex quadratic function, consider $f(x_1, x_2) = x_1^2 + x_2^2 + 6x_1 - 3x_2 + 16$. 

Note that this function can be equivalently written in the standard form as: 

$f(\mathbf{x}) = \frac{1}{2} {\mathbf{x}}^\top A \mathbf{x} +  {\mathbf{b}}^\top \mathbf{x} + c$ where $\mathbf{x} = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$, and

$
A = \begin{pmatrix}
2 & 0 \\ 0 & 2
\end{pmatrix}
$, $\mathbf{b} = \begin{pmatrix} 6 \\ -3 \end{pmatrix}$ and $c=16$.

On equating the gradient $\nabla f(\mathbf{x}) = \mathbf{0}$, we get $\begin{pmatrix} 2x_1 + 6 \\ 2x_2 - 3\end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$. 

Hence the point $\tilde{\mathbf{x}} = \begin{pmatrix} {\tilde{x}}_1 \\  {\tilde{x}}_2 \end{pmatrix} = \begin{pmatrix} -3 \\ \frac{3}{2} \end{pmatrix}$ satisfies $\nabla f(\tilde{\mathbf{x}}) = \mathbf{0}$.

Since $A$ is positive definite (eigen values are $2,2$), we have $\tilde{\mathbf{x}}$ as a local minimizer.

#A convex quadratic function with no local or global minimizer

Consider $f(x_1, x_2) = x_1^2 - 4x_2$. 

Note that this function can be equivalently written in the standard form as: 

$f(\mathbf{x}) = \frac{1}{2} {\mathbf{x}}^\top A \mathbf{x} +  {\mathbf{b}}^\top \mathbf{x} + c$ where $\mathbf{x} = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$, and

$
A = \begin{pmatrix}
2 & 0 \\ 0 & 0
\end{pmatrix}
$, $\mathbf{b} = \begin{pmatrix} 0 \\ -4 \end{pmatrix}$ and $c=0$.

On equating the gradient $\nabla f(\mathbf{x}) = \mathbf{0}$, we get $\begin{pmatrix} 2x_1  \\  -4\end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$, which does not have a solution for any value of $x_2$. 

Hence for any point $\mathbf{x} =\begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$, $\nabla f(\mathbf{x})$ can never be $\mathbf{0}$.

Let us check the graph of this function.

Similar to the one dimension case, we have the following necessary and sufficient conditions for minimality in higher dimensions. 



#First order necessary condition. 

Consider the optimization problem $\min\limits_{\mathbf{x} \in {\mathbb{R}}^n} f(\mathbf{x})$, where $f$ is continuously differentiable over ${\mathbb{R}}^n$. Let ${\mathbf{x}}^\star$ be a local minimizer of $f$, that is, $f({\mathbf{x}}^\star) \leq f(\mathbf{x}), \ \forall \mathbf{x} \in \mathcal{N}({\mathbf{x}}^\star, \epsilon)\subseteq {\mathbb{R}}^n$, for some $\epsilon > 0$. Then $\nabla f({\mathbf{x}}^\star) = \mathbf{0}$. 



#Stationary point

Let $f:{\mathbb{R}}^n\to \mathbb{R}$ be a continuously differentiable function.  A point ${\mathbf{x}}^\star \in {\mathbb{R}}^n$ is called a ${\color{blue}{\text{stationary point}}}$ of $f$ if $\nabla f ({\mathbf{x}}^\star) = \mathbf{0}$. 

________

Similar to one dimension case, the gradient condition $\nabla f(\tilde{{\mathbf{x}}}) = \mathbf{0}$ does not imply that $\tilde{\mathbf{x}}$ is a minimizer of $f$.

#Second order sufficient condition. 

Consider the optimization problem $\min\limits_{\mathbf{x} \in {\mathbb{R}}^n} f(\mathbf{x})$, where $f$ is twice continuously differentiable over ${\mathbb{R}}^n$. Let ${\mathbf{x}}^\star$ be a stationary point of $f$. If $\nabla^2 f({\mathbf{x}}^\star) \succ 0$, then ${\mathbf{x}}^\star$ is a $\textbf{strict local minimizer}$ of $f$, that is, $f({\mathbf{x}}^\star) < f(\mathbf{x}), \ \forall \mathbf{x} \in \mathcal{N}({\mathbf{x}}^\star, \epsilon)\subseteq {\mathbb{R}}^n$, for some $\epsilon > 0$.

___


Before we prove these conditions, we define a crucial extension of the derivative idea of one dimension to higher dimensions. 

#Directional derivative of a function $f$ at $\mathbf{x}$ along direction $\mathbf{d}$ 

Consider a function $f:{\mathbb{R}}^n \to \mathbb{R}$ and a point $\mathbf{x} \in {\mathbb{R}}^n$. Let $\mathbf{d} \in {\mathbb{R}}^n$ and let $\mathbf{d} \neq \mathbf{0}$. The directional derivative of $f$ at $\mathbf{x}$ along the non-zero direction $\mathbf{d}$ is defined as:

$
f'(\mathbf{x}; \mathbf{d}) = \lim\limits_{\alpha \downarrow 0} \frac{f(\mathbf{x}+\alpha \mathbf{d})-f(\mathbf{x})}{\alpha}. 
$

The directional derivative gives a notion of instantaneous rate of change of function value of $f$ when an infinitesimal movement is made from point $\mathbf{x}$ along the non-zero direction $\mathbf{d}$. 

___

#Relation between directional derivative and gradient 


Consider a function $f:{\mathbb{R}}^n \to \mathbb{R}$ and a point $\mathbf{x} \in {\mathbb{R}}^n$. Let $\mathbf{d} \in {\mathbb{R}}^n$ and let $\mathbf{d} \neq \mathbf{0}$. Then the directional derivative of $f$ at $\mathbf{x}$ along the non-zero direction $\mathbf{d}$ is related to the gradient $\nabla f(\mathbf{x})$ of $f$ at $\mathbf{x}$ as: 

$
f'(\mathbf{x}; \mathbf{d}) = \left \langle \nabla f(\mathbf{x}), \mathbf{d} \right \rangle. 
$
___

#Proof of the necessary condition. 

Consider $\mathbf{d} = {\mathbf{e}}_i = ( 0  \ldots \ 0 \  \underbrace{1}_{i-\text{th coordinate}} \ 0 \ \ldots \ 0 )^\top$. Note that ${\mathbf{e}}_i$ has a $1$ at $i$-th coordinate and $0$ at all other coordinates.  

Recall that ${\mathbf{x}}^\star$ is a minimizer of $f$. 

Let us consider the scalar function $g(\mu) := f({\mathbf{x}}^\star+\mu {\mathbf{e}}_i)$ . 

${\mathbf{x}}^\star$ is a minimizer of $f$ holds if and only if $0$ is minimizer of $g$, that is $g(0) \leq g(\mu) \ \forall \mu \in \mathbb{R}$. 

Then from necessary condition of optimality in one dimensions, we have $g'(0) = 0$. 

Now: 
\begin{align}
g'(0) &= \lim\limits_{\mu \downarrow 0} \frac{f({\mathbf{x}}^\star+\mu {\mathbf{e}}_i) - f(\mathbf{x})}{\mu} \\
&= f'({\mathbf{x}}^\star; {\mathbf{e}}_i) \\
&= \left \langle \nabla f({\mathbf{x}}^\star), {\mathbf{e}}_i \right \rangle  \\
&= \frac{\partial f({\mathbf{x}}^\star)}{\partial x_i}.
\end{align}

Since $g'(0)=0$ we have the partial derivative $\frac{\partial f({\mathbf{x}}^\star)}{\partial x_i} = 0$. 

Since the above analysis is for an arbitrary coordinate $i$, we can apply this analysis for every coordinate and hence for every coordinate $i$, the partial derivative  $\frac{\partial f({\mathbf{x}}^\star)}{\partial x_i} = 0$, which implies that $\nabla f({\mathbf{x}}^\star) = \mathbf{0}$.


Before we prove the sufficient condition, we state the linear approximation theorem for high dimensions.

#Line segment between two points 

Consider two points $\mathbf{x}, \mathbf{y} \in {\mathbb{R}}^n$. Then the line segment between two points $\mathbf{x}, \mathbf{y}$ is given by:

$
[\mathbf{x}, \mathbf{y}] = \lambda \mathbf{x} + (1-\lambda) \mathbf{y}, \ \forall \lambda \in [0,1].
$



#Linear Approximation Theorem. 

Let $f:{\mathbb{R}}^n \to \mathbb{R}$ be a twice continuously differentiable function over ${\mathbb{R}}^n$. Let $\mathbf{x} \in {\mathbb{R}}^n$ and let $r>0$ be such that the open interval $\mathcal{N}(\mathbf{x},r) \subseteq {\mathbb{R}}^n$. Then for any $\mathbf{y} \in \mathcal{N}(\mathbf{x},r)$, there exists ${\mathbf{\zeta}}_\mathbf{y} \in [\mathbf{x}, \mathbf{y}]$ such that: 

$
f(\mathbf{y}) {\color{red}{\mathbf{=}}} f(\mathbf{x}) + \nabla f(\mathbf{x})(\mathbf{y}-\mathbf{x}) + \frac{1}{2}(\mathbf{y}-\mathbf{x})^\top \nabla^2 f({\mathbf{\zeta}}_{\mathbf{y}})(\mathbf{y}-\mathbf{x}).  
$

#Proof of Second Order Sufficient Condition

Since $\nabla^2 f({\mathbf{x}}^\star) \succ 0$ and since $\nabla^2 f$ is continuous, there exists some $r>0$ such that $\nabla^2 f(\mathbf{x}) \succ 0, \ \forall \mathbf{x} \in \mathcal{N}({\mathbf{x}}^\star,r)$. That is, $\nabla^2 f(\mathbf{x})$ is positive definite for every point in a neighborhood of ${\mathbf{x}}^\star$. 

From the linear approximation theorem, we have at ${\mathbf{x}}^\star$ and for every ${\mathbf{y}} \in \mathcal{N}({\mathbf{x}}^\star,r)$, there exists ${\mathbf{\zeta}}_{\mathbf{y}} \in [\mathbf{x}, \mathbf{y}]$ such that: 

$
f(\mathbf{y}) {\color{red}{\mathbf{=}}} f({\mathbf{x}}^\star) + \nabla f({\mathbf{x}}^\star)(\mathbf{y}-{\mathbf{x}}^\star) + \frac{1}{2}(\mathbf{y}-{\mathbf{x}}^\star)^\top \nabla^2 f({\mathbf{\zeta}}_{\mathbf{y}})(\mathbf{y}-{\mathbf{x}}^\star).  
$

We further know that $\nabla f({\mathbf{x}}^\star)=\mathbf{0}$ and $\nabla^2 f(\mathbf{\zeta}_{\mathbf{y}}) \succ 0$. Thus:

$
f(\mathbf{y}) {\color{red}{\mathbf{=}}} f({\mathbf{x}}^\star) + \frac{1}{2}(\mathbf{y}-{\mathbf{x}}^\star)^\top \nabla^2 f({\mathbf{\zeta}}_{\mathbf{y}}) (\mathbf{y}-{\mathbf{x}}^\star)\implies f(\mathbf{y}) > f({\mathbf{x}}^\star).  
$

Hence ${\mathbf{x}}^\star$ is a strict local minimizer of $f$.
